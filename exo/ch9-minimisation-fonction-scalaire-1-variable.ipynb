# Minimisation de fonction 1D

Pour cet exemple, on peut voir que les algorithmes peuvent rester « coincés » dans des minimums locaux en fonction des paramètres fournis.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
import matplotlib
import matplotlib.pyplot as plt
import scipy
import scipy.stats as stats
import numpy as np
import scipy.optimize as optimize


def func_start(x):
    """
    x abscisses
    """
    return 1e-3 * x**4 - 1e-3 * x**3 - 4 * np.sin(4 * x + np.pi) + 4 * np.sin(x) + x


# Programme principal
if __name__ == "__main__":
    fig = plt.figure(figsize=(15, 8))
    gs = fig.add_gridspec(2, 2)
    ax1 = fig.add_subplot(gs[:, 0])
    x = np.linspace(-5 * np.pi, 5 * np.pi, 500)
    ax1.plot(x, func_start(x))

    # série des méthodes que l'on va utiliser
    method_list = ["Bounded", "Brent", "Brent", "Brent", "Brent"]
    # Intervalles ou bornes utilisées
    brackets = np.array(
        [
            [-5 * np.pi, 5 * np.pi],
            [-5 * np.pi, 5 * np.pi],
            [-10, 0],
            [-10, -7],
            [-10, -4],
        ]
    )

    for i, method in enumerate(method_list):
        if method == "Bounded":
            res = optimize.minimize_scalar(
                func_start, bounds=(brackets[i, 0], brackets[i, 1]), method="Bounded"
            )
        else:
            res = optimize.minimize_scalar(
                func_start, bracket=(brackets[i, 0], brackets[i, 1]), method=method
            )
        xsol = res.x
        print(
            "{} [{:.4f},{:.4f}] sol : x={}".format(
                method, brackets[i, 0], brackets[i, 1], xsol
            )
        )
        ax1.scatter(
            xsol,
            func_start(xsol),
            label="{} [{:.4f},{:.4f}]".format(method, brackets[i, 0], brackets[i, 1]),
        )
    ax1.legend()

    plt.show()
```

# Minimisation de fonction 1D de plusieurs variables

Encore une fois, en fonction du point de départ, les algorithmes ne convergent pas tous vers le minimum global de la fonction. Surtout quand on commence proche d'un minimum local, on a tendance à finir dans le minimum local correspondant.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
import matplotlib
import matplotlib.pyplot as plt
import scipy
import scipy.stats as stats
import numpy as np
import scipy.optimize as optimize
import matplotlib.cm as cm


def func_start(x, y):
    """
    x,y  abscisses
    function somme de plusieurs gaussienes centrées sur 5 positions dont on va chercher le minimum
    """
    return (
        -scipy.stats.norm.pdf(x, loc=0, scale=1)
        * scipy.stats.norm.pdf(y, loc=0, scale=1)
        + -2
        * scipy.stats.norm.pdf(x, loc=2, scale=1)
        * scipy.stats.norm.pdf(y, loc=2, scale=1)
        + -scipy.stats.norm.pdf(x, loc=-2, scale=1)
        * scipy.stats.norm.pdf(y, loc=-2, scale=1)
        + -scipy.stats.norm.pdf(x, loc=-0.5, scale=1)
        * scipy.stats.norm.pdf(y, loc=-0.5, scale=1)
        + -3
        * scipy.stats.norm.pdf(x, loc=2.5, scale=1)
        * scipy.stats.norm.pdf(y, loc=-2.5, scale=1)
    )


def func_mono(v):
    """
    v vecteur contenant les valeurs de xx et yy
    """
    x, y = v
    return (
        -scipy.stats.norm.pdf(x, loc=0, scale=1)
        * scipy.stats.norm.pdf(y, loc=0, scale=1)
        + -2
        * scipy.stats.norm.pdf(x, loc=2, scale=1)
        * scipy.stats.norm.pdf(y, loc=2, scale=1)
        + -scipy.stats.norm.pdf(x, loc=-2, scale=1)
        * scipy.stats.norm.pdf(y, loc=-2, scale=1)
        + -scipy.stats.norm.pdf(x, loc=-0.5, scale=1)
        * scipy.stats.norm.pdf(y, loc=-0.5, scale=1)
        + -3
        * scipy.stats.norm.pdf(x, loc=2.5, scale=1)
        * scipy.stats.norm.pdf(y, loc=-2.5, scale=1)
    )


def func_jac(v):
    x, y = v
    return x * scipy.stats.norm.pdf(x, loc=0, scale=1) * scipy.stats.norm.pdf(
        y, loc=0, scale=1
    ) + +(x - 2) * 2 * scipy.stats.norm.pdf(x, loc=2, scale=1) * scipy.stats.norm.pdf(
        y, loc=2, scale=1
    ) + +(
        x + 2
    ) * scipy.stats.norm.pdf(
        x, loc=-2, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=-2, scale=1
    ) + +(
        x + 0.5
    ) * scipy.stats.norm.pdf(
        x, loc=-0.5, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=-0.5, scale=1
    ) + +3 * (
        x - 2.5
    ) * scipy.stats.norm.pdf(
        x, loc=2.5, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=-2.5, scale=1
    ), y * scipy.stats.norm.pdf(
        x, loc=0, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=0, scale=1
    ) + +(
        y - 2
    ) * 2 * scipy.stats.norm.pdf(
        x, loc=2, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=2, scale=1
    ) + +(
        y + 2
    ) * scipy.stats.norm.pdf(
        x, loc=-2, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=-2, scale=1
    ) + +(
        y + 0.5
    ) * scipy.stats.norm.pdf(
        x, loc=-0.5, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=-0.5, scale=1
    ) + +3 * (
        y + 2.5
    ) * scipy.stats.norm.pdf(
        x, loc=2.5, scale=1
    ) * scipy.stats.norm.pdf(
        y, loc=-2.5, scale=1
    )


# Programme principal
if __name__ == "__main__":
    fig = plt.figure(figsize=(15, 8))
    gs = fig.add_gridspec(2, 2)
    ax1 = fig.add_subplot(gs[:, 0])
    x = np.linspace(-4, 4, 500)
    y = np.linspace(-4, 4, 500)
    xx, yy = np.meshgrid(x, y, indexing="xy")
    zz = func_start(xx, yy)

    ax1.imshow(
        zz, interpolation="bilinear", cmap=cm.bwr, origin="lower", extent=[-4, 4, -4, 4]
    )
    ax1.set_xlabel("x")
    ax1.set_ylabel("y")

    ax2 = fig.add_subplot(gs[:, 1], projection="3d")
    ax2.plot_surface(xx, yy, zz, cmap=cm.bwr, linewidth=0, antialiased=False)

    for starting_point in [[-1, 1], [0, 0], [2, -2], [4, -4]]:
        print("\n\nstarting point :  {}".format(starting_point))
        for method_name in [
            "Nelder-Mead",
            "Powell",
            "CG",
            "BFGS",
            "Newton-CG",
            "L-BFGS-B",
            "TNC",
            "COBYLA",
            "SLSQP",
            "trust-constr",
        ]:
            res = optimize.minimize(
                func_mono, starting_point, method=method_name, jac=func_jac
            )
            # print(res)
            if "nit" not in res:
                res.nit = -1
            print(
                "{:>15} [{:.4f},{:.4f}] sol : f(x)={} nb itérations {}".format(
                    method_name, res.x[0], res.x[1], func_mono(res.x), res.nit
                )
            )

    plt.show()
```
